```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
import itertools
import time
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsClassifier
import os
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest, chi2

%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.head()
data =  df['koi_disposition']!='CANDIDATE'
data.head()
```

```python
data2 = df[data]
data2.head()
```

```python
y = data2.koi_disposition
y_dummies = pd.get_dummies(y)
target_names = ["CONFIRMED", "FALSE POSITIVE"]
X = data2[['koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', \
          'koi_insol', 'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec', \
          'koi_kepmag']]
y_dummies = y_dummies.drop(['FALSE POSITIVE'], axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_dummies, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train.head()
```

```python
X_new = SelectKBest(chi2, k=10).fit_transform(X_train, y_train)
X_new[0:5]
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
print('Test Acc: %.3f' % model.score(X_train_scaled, y_train))
```

```python
print('Test Acc: %.3f' % model.score(X_test_scaled, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
from sklearn.metrics import confusion_matrix,plot_confusion_matrix
cm = confusion_matrix(y_test,predictions)
cm
```

```python
cmfig = plot_confusion_matrix(model,X_test_scaled,y_test)
```

```python
def fit_SVG(X,Y):
    #Fit support vector model and return RSS and R squared values
    model_k = SVC(kernel='linear')
    model_k.fit(X, Y)
    RSS = mean_squared_error(Y,model_k.predict(X)) * len(Y)
    R_squared = model_k.score(X,Y)
    return RSS, R_squared
```

```python
#Importing tqdm for the progress bar
from tqdm import tnrange, tqdm_notebook

#Initialization variables
Y = y_dummies
X = X

k = 40
RSS_list, R_squared_list, feature_list = [],[], []
numb_features = []

#Looping over k = 1 to k = 40 features in X
for k in tnrange(1,len(X.columns) + 1, desc = 'Loop...'):

    #Looping over all possible combinations: from 40 choose k
    for combo in itertools.combinations(X.columns,k):
        tmp_result = fit_SVG(X[list(combo)],Y)    
        RSS_list.append(tmp_result[0])                  #Append lists
        R_squared_list.append(tmp_result[1])
        feature_list.append(combo)
        numb_features.append(len(combo))   

#Store in DataFrame
df = pd.DataFrame({'numb_features': numb_features,'RSS': RSS_list, 'R_squared':R_squared_list,'features':feature_list})
```
